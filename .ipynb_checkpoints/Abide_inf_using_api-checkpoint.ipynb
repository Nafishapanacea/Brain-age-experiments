# Inference

```python
import requests
import json

model_path = 'http://182.77.48.121:8001/predict'
```

```python
import os

data_dir= '/home/jupyter-nafisha/Brain_age/Abide_IXI_data'
files= os.listdir(data_dir)
```

```python
files
```

```python
def predict(model_path, input_file):
    with open(input_file, "rb") as file:
        response = requests.post(model_path, files={"file": file})
        
        if response.status_code == 200:
            return response.content
            # print(response.content)
        else:
            print(f"Failed. Status code: {response.status_code}")
```

```python
for file in files:
    file_path= os.path.join(data_dir, file)
    # print(os.path.exists(file_path))
    output= predict(model_path, file_path)
    print(file,output)
```

```python
image_path = '/home/jupyter-nafisha/Brain_age/Temp_output/IXI079_outWarped.nii.gz'
output= predict(model_path, image_path)
```

```python
import json
import os
import numpy as np

count = 0 

actual_list = []
pred_list = []

for itr, row in df.iterrows():
    path = os.path.join(data_dir, row['relative_path'])
    actual_age = row['age_at_scan']
    
    model_response = predict(model_path, path)
    output_str = model_response.decode("utf-8")
    data = json.loads(output_str)
    predicted_age = data["predicted_brain_age"]
    
    print("actual_age:", actual_age, ", predicted age:", predicted_age)

    actual_list.append(actual_age)
    pred_list.append(predicted_age)

    # print(os.path.exists(path))

    count+=1
    if count%50 ==0:
        print("******Files processed: ",count)
        # break

# Calculate MAE 
actual_array = np.array(actual_list, dtype=float)
pred_array   = np.array(pred_list, dtype=float)

mae = np.mean(np.abs(actual_array - pred_array))

print("\nMean Absolute Error (MAE):", mae)
```
