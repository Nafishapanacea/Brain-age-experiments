```python
# import the csv

import pandas as pd
csv= pd.read_csv('/home/jupyter-nafisha/Brain_age/CSVs/Yale_orig.csv')
csv
```

```python
csv = csv[csv.groupby('patient_id')['patient_id'].transform('count') > 1]
csv
```

```python
csv['study_datetime'] = csv['study_datetime'].str.split('_').str[0]
csv
```

```python
import os
import pandas as pd

base_path = '/home/common/Yale-Brain-Mets-Longitudinal'

# Ensure new columns exist
if 'T1_pre_paths' not in csv.columns:
    csv['T1_pre_paths'] = None
if 'T1_post_paths' not in csv.columns:
    csv['T1_post_paths'] = None

# Traverse patient folders
for patient_id in sorted(os.listdir(base_path)):
    patient_path = os.path.join(base_path, patient_id)
    if not os.path.isdir(patient_path):
        continue
    if not os.path.exists(patient_path):
        print("patient record does not exists", patient_path)

    # Traverse study folders
    for study_folder in sorted(os.listdir(patient_path)):
        study_path = os.path.join(patient_path, study_folder)
        if not os.path.isdir(study_path):
            continue
        if not os.path.exists(study_path):
            print("Study does not exist", study_path)

        # Extract study date (first part before underscore)
        study_date = study_folder.split('_')[0]
        # print(study_date)

        # Collect T1 pre and post MRIs
        t1_pre_files, t1_post_files = [], []

        for f in os.listdir(study_path):
            # print(f)
            f_upper = f.upper()
            full_path = os.path.join(study_path, f)
            if 'PRE' in f_upper:
                t1_pre_files.append(full_path)
            elif 'POST' in f_upper:
                t1_post_files.append(full_path)

        # Update DataFrame only if the patient/date exists
        mask = (csv['patient_id'] == patient_id) & (csv['study_datetime'] == study_date)
        # print(mask)
        if mask.any():
            csv.loc[mask, 'T1_pre_paths'] = ', '.join(t1_pre_files) if t1_pre_files else None
            csv.loc[mask, 'T1_post_paths'] = ', '.join(t1_post_files) if t1_post_files else None
```

```python
csv['T1_pre_paths'] = csv['T1_pre_paths'].replace('None', pd.NA)

# Find patients for which ALL T1_pre_paths are NaN
patients_to_remove = (
    csv.groupby('patient_id')['T1_pre_paths']
    .apply(lambda x: x.isna().all())
)

# Keep only those patients where not all are NaN
df_filtered = csv[~csv['patient_id'].isin(patients_to_remove[patients_to_remove].index)]
```

```python
#  patients with all T1-MRIs

patients_with_missing = df_filtered.loc[df_filtered['T1_pre_paths'].isna() | (df_filtered['T1_pre_paths'].astype(str).str.strip() == ''), 'patient_id'].unique()
df_cleaned = df_filtered[~df_filtered['patient_id'].isin(patients_with_missing)]
```
