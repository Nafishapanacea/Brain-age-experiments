```python
import os
import shutil
import random

image_dir = '/home/common/IXI_synthetic_MRIs'
dest_dir  = '/home/jupyter-nafisha/Brain_age/IXI_synthetic_MRIs'

os.makedirs(dest_dir, exist_ok=True)

# collect nifti files
niftis = [
    f for f in os.listdir(image_dir)
    if f.endswith(('.nii', '.nii.gz'))
]

# randomly select 50 (or fewer if less exist)
selected = random.sample(niftis, min(50, len(niftis)))

for fname in selected:
    src = os.path.join(image_dir, fname)
    dst = os.path.join(dest_dir, fname)
    shutil.copy2(src, dst)

print(f"Copied {len(selected)} NIfTI files to {dest_dir}")
```
