```python
import os
import pandas as pd
data_path= '/home/omen/Downloads/oasis_cross-sectional_disc1/disc1'
xls_path= '/home/omen/Downloads/oasis_cross-sectional-5708aa0a98d82080.xlsx'
xls= pd.read_excel(xls_path)
xls
```

```python
new_csv= xls[['ID', 'Age', 'M/F']]
```

```python
valid_rows = []
for itr, row in new_csv.iterrows():
    path = os.path.join(data_path, row['ID'])
    if os.path.exists(path):
        valid_rows.append(True)
    else:
        valid_rows.append(False)
        print(f"Missing folder for ID: {row['ID']}")

# Filter DataFrame
new_csv = new_csv[valid_rows].reset_index(drop=True)
```

```python
new_csv
```

```python
import nibabel as nib

new_csv['file_path'] = None

for itr, row in new_csv.iterrows():
    input_dir= os.path.join(data_path, row['ID'], 'RAW')
    # print(input_dir)
    output_dir= os.path.join('/home/omen/Documents/Nafisha/brainAge/OASIS1_data',row['ID'])
    os.makedirs(output_dir, exist_ok=True)

    for file in os.listdir(input_dir):
        if file.endswith(".hdr"):
            base = os.path.splitext(file)[0]
            hdr_path = os.path.join(input_dir, base + ".hdr")
            img_path = os.path.join(input_dir, base + ".img")
            
            if os.path.exists(img_path):
                # Load the Analyze file
                img = nib.load(hdr_path)
                
                # Define output filename (.nii.gz)
                out_path = os.path.join(output_dir, base + ".nii.gz")
                
                # Save as NIfTI
                nib.save(img, out_path)
                print(f"Converted: {base} → {out_path}")
            else:
                print(f"⚠️ Missing .img for {base}")
    new_csv.at[itr, 'file_path'] = out_path
    
```

```python
new_csv.to_csv('/home/omen/Documents/Nafisha/brainAge/CSVs/OASIS_preprocessing.csv', index=False)
```

```python
import os
import pandas as pd
import shutil

# Paths
csv_path = '/home/omen/Documents/Nafisha/brainAge/CSVs/OASIS_preprocessing.csv'
dest_root = '/home/omen/Documents/Nafisha/brainAge/OASIS1_data'

# Read the CSV
df = pd.read_csv(csv_path)

# Make sure destination folder exists
os.makedirs(dest_root, exist_ok=True)

# Iterate and move files
for itr, row in df.iterrows():
    src = row['file_path']

    if not os.path.exists(src):
        print(f"⚠️ File not found: {src}")
        continue

    filename = os.path.basename(src)
    dest = os.path.join(dest_root, filename)

    try:
        shutil.move(src, dest)   # change to shutil.copy if you prefer copying
        df.at[itr, 'file_path'] = dest  # update path in CSV
        print(f"✅ Moved and updated path: {filename}")
    except Exception as e:
        print(f"❌ Error moving {src} → {dest}: {e}")


```
