```python
import os
import shutil

source_dir = '/home/common'
dest_dir = '/home/jupyter-nafisha/Brain_age/OASIS3'

os.makedirs(dest_dir, exist_ok=True)

for itr, row in data.iterrows():
    src_path = os.path.join(source_dir, row['file_path'])
    
    if not os.path.exists(src_path):
        print(f"File not found: {src_path}")
        continue
    
    file_name = os.path.basename(src_path)

    dst_path = os.path.join(dest_dir, file_name)

    # If filename already exists, add unique suffix
    if os.path.exists(dst_path):
        name, ext = os.path.splitext(file_name)
        new_name = f"{name}_{itr}{ext}"
        dst_path = os.path.join(dest_dir, new_name)

    # Copy file
    shutil.copyfile(src_path, dst_path)
    print(f"Copied: {src_path} → {dst_path}")
```
