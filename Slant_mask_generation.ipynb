```python
import requests
import os
import time
from pathlib import Path

API_URL = "http://localhost:8282"
OUTPUT_ROOT = "/home/omen/Documents/Nafisha/brainAge"

os.makedirs(OUTPUT_ROOT, exist_ok=True)

# image = row['image_path']
image_path = '/home/omen/Nikhil/AssemblyNet/testinginput/afterthought_1.nii.gz'

if not os.path.exists(image_path):
    print(f"Image not found: {image_path}")
#     continue

# --- Submit Job to API ---
print(f"Submitting to API: {image_path}")
with open(image_path, 'rb') as f:
    files = {'files': (os.path.basename(image_path), f, 'application/octet-stream')}
    try:
        response = requests.post(f"{API_URL}/slant_segment", files=files)
        response.raise_for_status()
    except Exception as e:
        print(f"Error submitting job: {e}")
#         continue

job_info = response.json()
job_id = job_info["job_id"]
print(f"Job submitted. ID: {job_id}")

# --- Poll for job completion ---
status_url = f"{API_URL}/job_status/{job_id}"
while True:
    time.sleep(10)
    status_resp = requests.get(status_url)
    if status_resp.status_code != 200:
        print("Failed to fetch job status")
        break

    status = status_resp.json()["status"]
    print(f"Status: {status}")

    if status == "completed":
        break
    elif status == "failed":
        print("Segmentation failed.")
        break

# --- Download result ---
download_url = f"{API_URL}/slant_download/{job_id}"
output_zip_path = os.path.join('/home/omen/Documents/Nafisha/brainAge', "slant_results.zip")
output_zip_path = output_zip_path.replace("\\", "/")
os.makedirs(os.path.dirname(output_zip_path), exist_ok=True)

try:
    result = requests.get(download_url)
    if result.status_code == 200:
        with open(output_zip_path, "wb") as out_file:
            out_file.write(result.content)
        print(f"Results downloaded: {output_zip_path}")
    else:
        print(f"Failed to download results for job {job_id}")
except Exception as e:
    print(f"Error downloading result: {e}")
```
