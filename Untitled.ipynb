```python
import pydicom

def is_mprage(path):
    d = pydicom.dcmread(path, stop_before_pixels=True)

    TI = float(getattr(d, "InversionTime", 0))
    TR = float(getattr(d, "RepetitionTime", 0))

    # Rule 1: TI must exist and be > 400 ms
    if TI < 400:
        return False

    # Rule 2: TR must be > 1500 ms
    if TR < 1500:
        return False

    return True
```

```python
import nibabel as nib
import numpy as np

path = '/home/jupyter-nafisha/Brain_age/OASIS3/sub-OAS30020_ses-d0092_run-02_T1w.nii.gz'

# Load image
img = nib.load(path)
data = img.get_fdata()

# Flip axis 0 (L â†’ R)
data_flipped = data[::-1, :, :]

# Create a new affine (RAS, 1mm spacing)
affine = np.array([
    [-1, 0, 0, img.shape[0]],   # X flipped
    [ 0, 1, 0, 0],              # Y same
    [ 0, 0, 1, 0],              # Z same
    [ 0, 0, 0, 1]
], dtype=float)

# Build new NIfTI
new_img = nib.Nifti1Image(data_flipped, affine)

# Write affine into qform + sform
new_img.set_qform(affine, code=1)
new_img.set_sform(affine, code=1)

out_path = '/home/jupyter-nafisha/Brain_age/sub-OAS30020_ses-d0092_run-02_T1w_affineFixed.nii.gz'
nib.save(new_img, out_path)

print("Saved:", out_path)
print("New orientation:", nib.aff2axcodes(new_img.affine))
```

```python
import nibabel as nib
from nibabel.orientations import axcodes2ornt, ornt_transform, inv_ornt_aff

path = '/home/jupyter-nafisha/Brain_age/OASIS3/sub-OAS30020_ses-d0092_run-02_T1w.nii.gz'

img = nib.load(path)
data = img.get_fdata()
affine = img.affine

# Current orientation
current_ornt = axcodes2ornt(nib.aff2axcodes(affine))
target_ornt = axcodes2ornt(('R','A','S'))

# Compute required transformation
transform = ornt_transform(current_ornt, target_ornt)

# Apply transform to data
new_data = nib.apply_orientation(data, transform)

# Compute new affine
new_affine = affine @ inv_ornt_aff(transform, data.shape)

new_img = nib.Nifti1Image(new_data, new_affine)
new_img.set_qform(new_affine, code=1)
new_img.set_sform(new_affine, code=1)

out = '/home/jupyter-nafisha/Brain_age/sub-OAS30020_ses-d0092_run-02_T1w_RAS_final.nii.gz'
nib.save(new_img, out)

print("Old orientation:", nib.aff2axcodes(affine))
print("New orientation:", nib.aff2axcodes(new_affine))
print("Saved:", out)
```

```python
import nibabel as nib
import numpy as np

# Load image
img = nib.load('/home/jupyter-nafisha/Brain_age/OASIS3/sub-OAS30048_ses-d0983_run-02_T1w.nii.gz')
data = img.get_fdata()
old_affine = img.affine

data_reoriented = np.transpose(data, (2, 0, 1))
 
data_reoriented = data_reoriented[:, ::-1, :]

T = np.array([
    [0, 0, 1, 0],                              # new x = old z
    [0,-1, 0, data_reoriented.shape[1] - 1],   # new y = -old x
    [1, 0, 0, 0],                              # new z = old y
    [0, 0, 0, 1]
])
new_affine = old_affine @ np.linalg.inv(T)

new_img = nib.Nifti1Image(data_reoriented, new_affine)

new_img.header.set_qform(new_affine, code=1)
new_img.header.set_sform(new_affine, code=1)

nib.save(new_img, 'Orient_mri_RAS_5_2_1.nii.gz')
```

```python
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt

nii_path = '/home/jupyter-nafisha/Brain_age/OASIS3/sub-OAS30048_ses-d0983_run-02_T1w.nii.gz'

nii = nib.load(nii_path)
data = nii.get_fdata()

# print("Image shape:", data.shape)

x_mid = data.shape[0] // 2  # sagittal
y_mid = data.shape[1] // 2  # coronal
z_mid = data.shape[2] // 2  # axial

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Axial view (top-down)
axes[0].imshow(np.rot90(data[:, :, z_mid]), cmap='gray')
axes[0].axis('off')

# Sagittal view (left-right)
axes[1].imshow(np.rot90(data[x_mid, :, :]), cmap='gray')
axes[1].axis('off')

# Coronal view (front-back)
axes[2].imshow(np.rot90(data[:, y_mid, :]), cmap='gray')
axes[2].axis('off')

plt.tight_layout()
plt.show()
```
