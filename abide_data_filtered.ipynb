```python
csv_path= '/home/omen/Documents/Nafisha/brainAge/CSVs/Phenotypic_V1_0b.csv'
data_path='/home/omen/Documents/Megha/AUTISM_PREPROCESSED'
```

```python
import pandas as pd
df= pd.read_csv(csv_path)
# df
```

```python
# autism patients have DX_GROUP=1
autism_data= df[df['DX_GROUP']==1]
```

```python
autism_data['SUB_ID'] = autism_data['SUB_ID'].astype(str).str.strip()
```

```python
import os

autism_data['file_name'] = None   

for path in os.listdir(data_path):
    id = path[-12:-7].strip()
    autism_data.loc[autism_data['SUB_ID'] == id, 'file_name'] = path

# autism_data
```

```python
autism_data = autism_data[autism_data['file_name'].notna()]
autism_data
```

### removing  _harmonized

```python
data_dir= '/home/omen/Documents/Nafisha/brainAge/AUTISM_PREPROCESSED_with_CBH'
```

```python
import os

for fname in os.listdir(data_dir):
    if "_harmonized" in fname:
        old_path = os.path.join(data_dir, fname)
        new_name = fname.replace("_harmonized", "")
        new_path = os.path.join(data_dir, new_name)

        os.rename(old_path, new_path)
        print(f"Renamed: {fname} → {new_name}")
```
