```python
import os
import pandas as pd
processed_ABIDE2_Path = 'CSVs/CBH_processed_ABIDE2.csv'
preprocessed_IXI_Path = 'CSVs/CBH_preprocessed_IXI.csv'
preprocessed_ABIDE1_Path = 'CSVs/CBH_preprocessed_ABIDE1.csv'
processed_ABIDE2 = pd.read_csv(processed_ABIDE2_Path)
preprocessed_IXI = pd.read_csv(preprocessed_IXI_Path)
preprocessed_ABIDE1 = pd.read_csv(preprocessed_ABIDE1_Path)
ABIDE2_filenames = list(processed_ABIDE2['file name'])
IXI_filenames = list(preprocessed_IXI['file_name'])
ABIDE1_filenames = list(preprocessed_ABIDE1['file_name'])
```

```python
DataDir = '/home/jupyter-nafisha/Brain_age/Trimeese_Data_without_CBH'
DataFiles = os.listdir(DataDir)
```

```python
NotinCSV = []
InCSV = []
CombinedData = []
for file in DataFiles:
    if file in ABIDE2_filenames:
        InCSV.append(file)
        age = processed_ABIDE2[processed_ABIDE2['file name'] == file]['age'].values[0]
        sex = processed_ABIDE2[processed_ABIDE2['file name'] == file]['sex'].values[0]
        site = processed_ABIDE2[processed_ABIDE2['file name'] == file]['site'].values[0]
        CombinedData.append([file, age, sex, site])
        print(age)
    elif file in IXI_filenames:
        InCSV.append(file)
        age = preprocessed_IXI[preprocessed_IXI['file_name'] == file]['age'].values[0]
        sex = preprocessed_IXI[preprocessed_IXI['file_name'] == file]['sex'].values[0]
        site = preprocessed_IXI[preprocessed_IXI['file_name'] == file]['site'].values[0]
        CombinedData.append([file, age, sex, site])
    elif file in ABIDE1_filenames:
        InCSV.append(file)
        age = preprocessed_ABIDE1[preprocessed_ABIDE1['file_name'] == file]['age'].values[0]
        sex = preprocessed_ABIDE1[preprocessed_ABIDE1['file_name'] == file]['sex'].values[0]
        site = preprocessed_ABIDE1[preprocessed_ABIDE1['file_name'] == file]['site'].values[0]
        CombinedData.append([file, age, sex, site])
    else:
        NotinCSV.append(file)  
```

```python
CombinedDataFrame = pd.DataFrame(CombinedData, columns=['file_name', 'age', 'sex', 'site'])
CombinedDataFrame = CombinedDataFrame.dropna()

combined_filenames = list(CombinedDataFrame['file_name'])
```
