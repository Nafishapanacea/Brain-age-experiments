```python
import os

# Folder path
folder_path = "/home/omen/Documents/Megha/AUTISM_PREPROCESSED_with_CBH"

# Counter for deleted files
deleted = 0

# Loop through files in the folder
for root, dirs, files in os.walk(folder_path):
    for file in files:
        if "_stand_mean" in file:
            file_path = os.path.join(root, file)
            try:
                os.remove(file_path)
                deleted += 1
                print(f"Deleted: {file_path}")
            except Exception as e:
                print(f"Error deleting {file_path}: {e}")

print(f"\Total files deleted: {deleted}")
```

```python
import os

# Folder path
folder_path = "/home/omen/Documents/Nafisha/brainAge/Yale_preprocessed_data_with_CBH"

renamed = 0

# Loop through files
for root, dirs, files in os.walk(folder_path):
    for file in files:
        if "_harmonized" in file:
            old_path = os.path.join(root, file)
            new_name = file.replace("_harmonized", "")
            new_path = os.path.join(root, new_name)
            
            # Rename file
            os.rename(old_path, new_path)
            renamed += 1
            print(f"Renamed: {old_path} → {new_path}")

print(f"\nTotal files renamed: {renamed}")
```
