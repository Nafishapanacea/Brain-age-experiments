### data distribution for triamese training only on nonlinear registrtered data without  CBH

```python
import pandas as pd
data_path= '/home/omen/Documents/Megha/Trimeese_Data_without_CBH'
```

### training_data

```python
train_csv= '/home/omen/Documents/Rishabh/rp-ai-Brain-age-data-distribution/Balancedmeanandstd_dataset/TrainCSV.csv'
train_data= pd.read_csv(train_csv)
train_data
```

```python
import os
import shutil

dest_train= '/home/omen/Documents/Nafisha/brainAge/triamese_vit_without_CBH/NTrain'
for itr,row in train_data.iterrows():
    file_path= os.path.join(data_path, row['FILE_NAME'])
    if os.path.exists(file_path):
        shutil.copy(file_path, dest_train)
    else:
        print(f"File not found: {file_path}")
```

```python
# Testing
for itr,row in train_data.iterrows():
    file_path= os.path.join(dest_train, row['FILE_NAME'])
    if not os.path.exists(file_path):
        print(file_path)
```
