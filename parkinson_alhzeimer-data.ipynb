# making copy of parkinson data

```python
source_dir= '/home/omen/Documents/Megha/ADNI_PARKINSONS_WORK/parkinson_alzheimers_original_data/PARKINSONS_NIFTI'
dest_dir= '/home/omen/Documents/Nafisha/brainAge/parkinson_data'
csv_path= '/home/omen/Documents/Nafisha/brainAge/CSVs/PARKINSON_filtered.csv'
```

```python
import pandas as pd
park_data= pd.read_csv(csv_path)
```

```python
import os
import shutil

for itr, row in park_data.iterrows():
    file_path= os.path.join(source_dir, row['file_path'])
    dest_path= os.path.join(dest_dir, row['file_path'])
    shutil.copy(file_path,dest_path)
```
