# parkinson csv

```python
csv_path= '/home/jupyter-nafisha/Brain_age/CSVs/PARKINSON.csv'
```

```python
import pandas as pd
park_data= pd.read_csv(csv_path)
park_data
```

```python
park_data["file_path"] = (
    park_data["subject_id"].astype(str)
    + "_"
    + park_data["image_uid"].astype(str)
    + ".nii.gz"
)
```

```python
filtered_data= park_data.iloc[0:50]
```
